```python
sorted_rscore=sorted(df['Reading Score'])
```

```python
sorted_rscore
```

```python
q1=np.percentile(sorted_rscore,25)
q3=np.percentile(sorted_rscore,75)
print(q1,q3)
```

```python
IQR=q3-q1
IQR
```

```python
lwr_bound=q1-(1.5*IQR)
uper_bound=q3+(1.5*IQR)
print(lwr_bound,uper_bound)
```

```python
r_outliers=[]
for i in sorted_rscore:
    if(i>lwr_bound or i<uper_bound):
        r_outliers.append(i)
        print(r_outliers)
```
